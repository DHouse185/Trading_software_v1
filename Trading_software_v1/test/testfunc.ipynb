```python
import typing
class MyData:
    def __init__(self, data: typing.List):
        "Initialize MyData from a sequence"
        self.data = data
     
    @classmethod
    def fromfilename(cls, filename: int):
        "Initialize MyData from a file"
        data = filename
        return cls(data)
     
    @classmethod
    def fromdict(cls, datadict: typing.Dict):
        data = datadict
        return cls(data.items())
```

```python
TF_EQUIV = {"1m": 60, "5m": 300, "15m": 900, "30m": 1800, "1h": 3600, "4h": 14400}
DEFAULT_BACKTEST_PARAMETERS = {
    "Balance"         : { 'Default'     : 3000},
    "Balance_percent" : { 'Default'     : 0.02},
    "Take_Profit"     : { 'Default'     : 0.02},
    "Stop_Loss"       : { 'Default'     : 0.005},
    }
class BackTest:
    def __backtest__(self, data, parameters):
        base_default_params = DEFAULT_BACKTEST_PARAMETERS
        self.balance = base_default_params["Balance"]["Default"]
        self.balance_percentage = base_default_params["Balance_percent"]["Default"]
        self.take_profit = base_default_params["Take_Profit"]["Default"]
        self.stop_loss = base_default_params["Stop_Loss"]["Default"]
         
        for params in parameters:
            if "Balance" in params:
                self.balance = params["Balance"]
            if "Balance_percent" in params:
                self.balance_perc = params["Balance_percent"]
            if "Take_Profit" in params:
                self.take_profit = params["Take_Profit"]
            if "Stop_Loss" in params:
                self.stop_loss = params["Stop_Loss"] 
                
        parameter = [self.balance, self.balance_percentage, self.take_profit, self.stop_loss]
                
        return parameter   
    
    def _default_parameters():
        ...
        
class Strategy:
    """
    Class that will manage all strategies and their
    properties
    """
    def __init__(self, client: str, pair: typing.Dict, exchange: str, timeframe: str, balance_percentage: float,
                 take_profit: float, stop_loss: float, strat_name):
        # Initiate from passed values
        self.client = client
        self.pair = pair
        self.exchange = exchange
        self.timeframe = timeframe
        self.tf_equiv = TF_EQUIV[timeframe] * 100 
        self.balance_percentage = balance_percentage
        self.take_profit = take_profit
        self.stop_loss = stop_loss
        self.strat_name = strat_name
        # when this is initlized, positions are not
        # initialized
        self.ongoing_position = False
        # Initializing strategies properties 
        self.candles: typing.List = []
        self.trades: typing.List = []
        self.logs = []
        
class TechnicalStrategy(Strategy, BackTest):
    def __init__(self, 
                 client: str, 
                 pair: typing.Dict, 
                 exchange: str, 
                 timeframe: str,
                 balance_percentage: float, 
                 take_profit: float, 
                 stop_loss: float, 
                 other_params: typing.Dict):
        super().__init__(client, pair, exchange, timeframe, balance_percentage,
                 take_profit, stop_loss, "Technical") 
        # Initiate parameters for technical strategies
        self._ema_fast = other_params['ema_fast']
        self._slow_ema = other_params['ema_slow']
        self._ema_signal = other_params['ema_signal']
        self._rsi_length = other_params['rsi_length']
        
    @classmethod
    def backtesting(cls, data: pd.DataFrame, parameters: typing.List[typing.Dict]):
        data = cls.__backtest__(cls, data, parameters) 
        return data
```

```python
import numpy as np
import pylab as plt

np.random.seed(28)
N = 300
rets = np.random.randn(N) 
prices = np.cumsum(rets)+10
plt.plot(prices)
plt.ylabel("Prices")
```

```python
inpos = 0
pos = []
prx = []
pnl_real = []
e_time = []
mav = [3, 9]
```

```python
for i in range(max(mav) + 1, N):
    ma1 = np.mean(prices[i-mav[0]:i])
    ma2 = np.mean(prices[i-mav[1]:i])
    
    pos.append(inpos)
    prx.append(prices[1])
    
    if ma1 > ma2:
        if inpos != -1:
            if inpos:
                pnl_real.append(prices[i] - entry)
                e_time.append(i)
                
            entry = prices[i]
            inpos = -1
            
    else:
        if inpos != 1:
            if inpos:
                pnl_real.append(entry - prices[i])
                e_time.append(i)
                
            entry = prices[i]
            inpos = 1
```

```python
plt.plot(np.array(e_time) - max(mav) - 2, np.cumsum(pnl_real), '-o')
```
